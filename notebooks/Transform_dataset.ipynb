```python
annotatedd_graph_df = annotatedd_graph_df[annotatedd_graph_df['source'].isin(entity2id) & annotatedd_graph_df['target'].isin(entity2id)]
```

```python
annotated_graph_edges = np.zeros(annotatedd_graph_df.shape)
```

```python
annotated_graph_edges[:, 0] = annotatedd_graph_df['source'].map(entity2id)
```

```python
annotated_graph_edges[:, 2] = annotatedd_graph_df['target'].map(entity2id)
```

```python
unique_relations = annotatedd_graph_df['annotations'].unique().tolist()
```

```python
relations2id = {relation: i for i, relation in enumerate(unique_relations)}
id2relation = {i: relation for i, relation in enumerate(unique_relations)}
```

```python
annotated_graph_edges[:, 1] = annotatedd_graph_df['annotations'].map(relations2id)
```

```python
annotatedd_graph_df[~annotatedd_graph_df['source'].isin(entity2id)]['source'].unique()
```

```python
annotated_graph_edges = annotated_graph_edges.astype(int)
```

```python
user_relation = max(list(id2relation.keys())) + 1
```

```python
user_edges = np.zeros((interactions.shape[0], 3))
user_edges[:, 0] = interactions['user_id'].map(entity2id)
user_edges[:, 2] = interactions['course_name'].map(entity2id)
user_edges[:, 1] = user_relation
```

```python
user_edges = user_edges.astype(int)
```

```python
all_edges = np.concatenate([user_edges, annotated_graph_edges], axis=0)
```

```python
np.save('edges_gcf.npy', all_edges)
```
